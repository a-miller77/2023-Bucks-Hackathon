# Comparing Defenders against Nading by Height #

This Journal's main purpose is to take all of the data from Nading's Matchups and merging player's results by height into Taller, about the same, and shorter 

```python
#Created By Elias Vera-Jimenez

# Base Imports
import pandas as pd
import numpy as np
import calculate as calc

df = pd.read_csv('filtered_sets/only_nading_matchups.csv') # Accessing the CSV with only nading matchups 
```

```python
# Storing the names & height of all of the players that are taller than nading
taller_than_nading = df[df["player_defMatchup_ht"] > 81 ]

taller_than_nading

```

```python
# Storing the names & heights of all of the players that are around the same or the same height as Nading
around_nading_height = df[df["player_defMatchup_ht"].between(77,81)]

around_nading_height
```

```python
shorter_than_nading = df[df["player_defMatchup_ht"] < 77]

shorter_than_nading
```

```python
#list of all of the different statistics (not including: player names, heights, or possessions
player_stats_list = [
    "player_defMatchup_ht",
    "ptsScored_team", 
    "fga", 
    "fg", 
    "fga3", 
    "fg3", 
    "qSQ", 
    "iso_actions", 
    "iso_pts", 
    "pick_actions", 
    "pick_pts",
    "post_actions",
    "post_pts",
    "offBall_actions",
    "offBall_pts",
    "passes",
    "assistOppCreated",
    "touches",
    "touches_direct",
    "touches_catchAndShoot",
    "touches_paint",
    "touches_time",
    "tov",
    "fta",
    "ft_player",
    "fouls"
]
print(len(player_stats_list))

# Dictionary that will contain all of the data for the players
players_stats_avg_dict = {
    "def_heights" : ["Shorter Players", "Around-Height Players", "Taller Players"],
    "avg_def_height" : [],
    "avg_ptsScored_team" : [],
    "avg_fga" : [],
    "avg_fg": [], 
    "avg_fga3": [], 
    "avg_fg3" : [], 
    "avg_qSQ" : [], 
    "avg_iso_actions" : [], 
    "avg_iso_pts": [], 
    "avg_pick_actions": [], 
    "avg_pick_pts" : [],
    "avg_post_actions" : [],
    "avg_post_pts" : [],
    "avg_offBall_actions": [],
    "avg_offBall_pts" : [],
    "avg_passes" : [],
    "avg_assistOppCreated" : [],
    "avg_touches" : [] ,
    "avg_touches_direct" : [],
    "avg_touches_catchAndShoot" : [],
    "avg_touches_paint" : [],
    "avg_touches_time" :[] ,
    "avg_tov" : [],
    "avg_fta" : [],
    "avg_ft_player" : [] ,
    "avg_fouls": []
}

do_not_change_na = [
    "iso_pts",
    "iso_actions",
    "pick_pts",
    "pick_actions",
    "post_pts",
    "post_actions",
    "offBall_pts",
    "offBall_actions",
    "touches_time"
]
```

```python

def get_stats (df):
    avg_stats = []
    for stat in player_stats_list:
        if stat not in do_not_change_na:
            df.loc[:, f'{stat}'].fillna(0)
        avg_stats.append(np.round(np.nanmean(df[stat]), 3))
    return avg_stats

shorter_avgs = get_stats(shorter_than_nading)

taller_avgs = get_stats(taller_than_nading)

same_avgs = get_stats(around_nading_height)

```

```python
shorter_avgs = get_stats(shorter_than_nading)

taller_avgs = get_stats(taller_than_nading)

same_avgs = get_stats(around_nading_height)

print(len(shorter_avgs), len(taller_avgs), len(same_avgs))
print(same_avgs, shorter_avgs, taller_avgs)
index = 0
for stat in players_stats_avg_dict:
    if(stat == "def_heights"):
        players_stats_avg_dict[stat] = ["Shorter", "Same", "Taller"]
    else:
        players_stats_avg_dict[stat] = [shorter_avgs[index],same_avgs[index],taller_avgs[index]]
        index+=1
```

```python

heights_avg = pd.DataFrame.from_dict(players_stats_avg_dict)

heights_avg = calc.update_df(heights_avg)

heights_avg

#heights_avg.to_csv('filtered_sets/nading_vs_different_heights_averages.csv')
```
