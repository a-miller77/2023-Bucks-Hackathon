```python
#Created By Elias Vera-Jimenez

# Base Imports
import pandas as pd
import numpy as np

df = pd.read_csv('filtered_sets/only_nading_matchups.csv') # Accessing the CSV with only nading matchups 
```

```python
# Storing the names of all of the players that are taller than nading
taller_than_nading_raw = df[df["player_off_ht"] < df["player_defMatchup_ht"]]

taller_than_nading_raw = taller_than_nading_raw[["player_defMatchup_a","player_defMatchup_ht"]]

taller_than_nading = taller_than_nading_raw.drop_duplicates()

taller_than_nading
```

```python
# Storing the names of all of the players that are taller than nading
shorter_than_nading_raw = df[df["player_off_ht"] >= df["player_defMatchup_ht"]]

shorter_than_nading_raw = shorter_than_nading_raw[["player_defMatchup_a","player_defMatchup_ht"]]

shorter_than_nading = shorter_than_nading_raw.drop_duplicates()

shorter_than_nading
```
